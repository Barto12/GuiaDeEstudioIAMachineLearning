```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
import numpy as np # Import numpy

# Define or load your dataset and assign it to X_scaled
# Example with sample data (replace with your actual data loading)
X_scaled = np.random.rand(100, 5) # Creates a sample dataset of 100 samples and 5 features

# Calcular inercia para diferentes valores de K
inertias = []
for k in range(1, 10):
    kmeans = KMeans(n_clusters=k, random_state=42, n_init=10) # Added n_init to suppress future warnings
    kmeans.fit(X_scaled)
    inertias.append(kmeans.inertia_)

# Gráfico del método del codo
plt.plot(range(1, 10), inertias, marker='o')
plt.title("Método del codo")
plt.xlabel("Número de clusters (K)")
plt.ylabel("Inercia")
plt.grid()
plt.show()
```
