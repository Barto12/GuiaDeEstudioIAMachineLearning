```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler

# Paso 1: Cargar dataset
df = sns.load_dataset('iris')

# Paso 2: Revisar valores nulos
print("Valores nulos por columna:")
print(df.isnull().sum())

# Paso 3: Estad√≠sticas b√°sicas
print("\nResumen estad√≠stico:")
print(df.describe())

# Paso 4: Histograma general
df.hist(figsize=(10, 6))
plt.suptitle("Distribuci√≥n de caracter√≠sticas", fontsize=16)
plt.show()

# Paso 5: Heatmap de correlaciones
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap="coolwarm")
plt.title("Matriz de correlaci√≥n")
plt.show()

# Paso 6: Detecci√≥n visual de outliers con boxplot
sns.boxplot(data=df[['sepal_length', 'sepal_width', 'petal_length', 'petal_width']])
plt.title("Boxplot para detecci√≥n de outliers")
plt.xticks(rotation=45)
plt.show()

# Paso 7: Estandarizar variables num√©ricas
scaler = StandardScaler()
scaled = scaler.fit_transform(df.drop('species', axis=1))
scaled_df = pd.DataFrame(scaled, columns=df.columns[:-1])

print("\nVariables estandarizadas (primeras filas):")
display(scaled_df.head())
```
