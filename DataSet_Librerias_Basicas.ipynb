```python
# Paso 1: Importar librerías básicas
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Paso 2: Cargar dataset Iris desde seaborn
df = sns.load_dataset('iris')

# Paso 3: Mostrar las primeras 5 filas del dataset
print("Primeras filas del dataset Iris:")
display(df.head())

# Paso 4: Visualizar con un gráfico de dispersión
sns.scatterplot(data=df, x="sepal_length", y="sepal_width", hue="species")
plt.title("Dispersión entre largo y ancho de sépalo")
plt.show()
```
